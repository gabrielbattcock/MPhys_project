```python
def convertNRRD(filename, folders_patients):
    #for filename in folders_studyID:
        if filename.endswith('.seg.nrrd'):
            return False
        elif filename.endswith('.nrrd'):
            #print(filename)
            filepath=os.path.join(folders_patients,filename)
            readdata, header = nrrd.read(filepath)
            readdata=np.rot90(readdata,3)
            readdata=np.flip(readdata,1)

            return readdata
        else:
            return False

        
```
